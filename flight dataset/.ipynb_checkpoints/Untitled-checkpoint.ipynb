# Q.1 What are the airlines in the dataset, accompanied by their frequencies

```python
data.head()
```

```python
data['airline'].nunique() #checking number of airlines
```

```python
data['airline'].unique()
```

```python
data['airline'].value_counts() #checking their frequencies
```

```python
data['airline'].value_counts().sum() #total adds upto ....
```

```python
#Showing the airlines with their number of flights in a Bar graph
data['airline'].value_counts(ascending=True).plot.bar()
```

```python
plt.show()
```

```python
data['airline'].value_counts(ascending=True).plot.barh(color=['lightblue','lightgreen'])
plt.title("Airline with frequencies")
plt.xlabel("Number of flights")
plt.ylabel("airline")
plt.show()
```

# Q.2 Show the Bar Graphs representing the Departure Time and  the Arrival Time

```python
data.head(2)
```

```python
data['departure_time'].value_counts()
```

```python
data['arrival_time'].value_counts()
```

```python
plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
plt.bar(data['departure_time'].value_counts().index,data['departure_time'].value_counts().values,color=['r','b'])
plt.title("Departure Time")
plt.ylabel("Departure Frequency")
plt.xlabel("Departure Time")

plt.subplot(1,2,2)
plt.bar(data['arrival_time'].value_counts().index,data['arrival_time'].value_counts().values,color=['g','y'])
plt.title("ARRIVAL Time")
plt.ylabel("Arrival Frequency")
plt.xlabel("Arrival Time")

plt.show()
```

# Q.3 Show Bar Graphs representing the Source city and Destination City

```python
data.head(1)
```

```python
data['source_city'].value_counts()
```

```python
data['destination_city'].value_counts()
```

```python
plt.figure(figsize=(16,4))
plt.subplot(1,2,1)

plt.barh(data['source_city'].value_counts().index,data['source_city'].value_counts().values,color=['m','g'])
plt.title(" source Cities")
plt.xlabel('No of Flights')
plt.ylabel('cities')

plt.subplot(1,2,2)

plt.barh(data['destination_city'].value_counts().index,data['destination_city'].value_counts().values,color=['y','r'])
plt.title("Destination Cities")
plt.xlabel('No of Flights')
plt.ylabel('cities')
plt.show()
```

# Q.4 Does the price varies with the airlines?

```python
data.head(1)
```

```python
# Grouping the airlines and checking their mean prices
data.groupby('airline')['price'].mean()
```

```python
#Drawing a bar graph showing this output
sns.catplot(x='airline',y='price',kind='bar',palette="rocket",data=data,hue="class")
plt.show()
```

# Q.5 Does ticket price change based on the departure time and the arrival time

```python
data.head(2)
```

```python
# Mean ticket price based on the departure time
data.groupby('departure_time')['price'].mean()
```

```python
# Mean ticket price based on the arrival time
data.groupby('arrival_time')['price'].mean()
```

```python
sns.catplot(x='departure_time',y='price',kind='bar',palette="rocket",data=data)
plt.show()
```

```python
sns.catplot(x='arrival_time',y='price',kind='bar',palette="rocket",data=data)
plt.show()
```

```python
sns.relplot(x='arrival_time',y="price",data=data,col='departure_time',kind="line")
plt.show()
```
